**Descriptive Statistics - Mean**

Mean: To find mean of a number of items, the first step will be create a vector and then use the command mean()


```r
height <- c(100, 106, 121, 111, 109, 111, 103, 117, 114)
height
```

```r
mean(height)
```

Some times all values are not available. If there is no value for an item, it is stored as NA.

If there is an NA value in the vector, the mean will shown as NA.

```r
height1 <- c(100, 106, 121, 111, 109, 111, 103, 117, 114, NA, 1047)
height1
```

```r
mean(height1)
```

In case as above, you will need to remove NA from the calculation. That can be done by argument na.rm = TRUE


```r
height1 <- c(100, 106, 121, 111, 109, 111, 103, 117, 114, NA, 1047)
mean(height1, na.rm = TRUE)
```

In the above case there is an extreme value of 1047. In some cases you might want to remove these outliers from the calculation.

In R there is an option to trim the data. The argument trim = 0.1 (for example) will remove
10% data from the top and 10% from the bottom

```r
height1 <- c(100, 106, 121, 111, 109, 111, 103, 117, 114, NA, 1047)
mean(height1, trim=0.1, na.rm = TRUE)
```

```r
#Create 10,000 numbers with uniform distribution
random <- runif(10000)
hist(random)
```

**Check the mean and standard deviation of 10,000 population

```r
mean_random <- mean(random)
mean_random
```

```r
std_random <- sd(random)
std_random
```

**Draw four random samples from random and calculate mean**


```r
sample(random, 5)
```

```r
mean(sample(random, 5))
```

**Draw 5 samples from the uniformly distributed population "random", 1000 times and take
mean of these and draw the histogram of these sample means**


```r
random5 <- mean(sample(random,5))
for (i in 1:1000){
  random5 <- c(random5, mean(sample(random, 5)))
}
hist(random5)
```

**Check the mean and standard deviation, std of these sample means which have been
stored in the vector random5**


```r
mean_random5 <- mean(random5)
mean_random5
```

```r
std_random5 <- sd(random5)
std_random5
```

**Now instead of taking 5 samples at a time, lets draw 10 samples at a time.**

This will make the histogram of more narrow as the standard deviation of sample means will be standard deviation of individual items divided by square root of 10.


```r
random10 <- mean(sample(random, 10))

for (i in 1:1000){
  random10 <- c(random10, mean(sample(random, 10)))
}
hist(random10)
```

Check the mean and standard deviation of the means of 10 samples drawn at a time.

```r
mean_random10 <- mean(random10)
mean_random10
```

```r
std_random10 <- sd(random10)
std_random10
```

Repeat the sampel process with 100 samples drawn at a time.


```r

random100 <- mean(sample(random,100))
for (i in 1:1000){
  random100 <- c(random100, mean(sample(random, 100)))
}
hist(random100)
```

```r
mean_random100 <- mean(random10)
mean_random100
```

```r
std_random100 <- sd(random100)
std_random100
```

**Summary of Central Limit Theorem on a Sample Chart**

https://bookdown.org/ndphillips/YaRrr/arranging-plots-with-parmfrow-and-layout.html

```r
par(mfrow = c(2,2))

hist(random, main = "1 Sample")
mtext(std_random, side=3)
mtext(mean_random, side=4)

hist(random5, main = "5 Samples")
mtext(std_random5, side=3)
mtext(mean_random5, side=4)

hist(random10, main="10 samples")
mtext(std_random10, side=3)
mtext(mean_random10, side=4)

hist(random100, main="100 sample")
mtext(std_random100, side=3)
mtext(mean_random100, side=4)
```

We can use round() function and xlim to refine  the above visualisation

xlim: Convenience functions to set the limits of the x and y axis

https://www.rdocumentation.org/packages/ggplot2/versions/1.0.0/topics/xlim


https://www.statology.org/xlim-ylim-in-r/

```r
par(mfrow = c(2,2))

hist(random, main = "1 Sample", xlim = c(0, 1.0))
mtext(round(std_random, 2), side=3)
mtext(round(mean_random, 2), side=4)

hist(random5, main = "5 Samples", xlim = c(0, 1.0))
mtext(round(std_random5, 2), side=3)
mtext(round(mean_random5, 2), side=4)

hist(random10, main="10 samples", xlim = c(0, 1.0))
mtext(round(std_random10, 2), side=3)
mtext(round(mean_random10, 2), side=4)

hist(random100, main="100 sample", xlim = c(0, 1.0))
mtext(round(std_random100, 2), side=3)
mtext(round(mean_random100, 2), side=4)
```

**BINOMIAL PROBABILITY DISTRIBUTION**

https://www.geeksforgeeks.org/a-guide-to-dbinom-pbinom-qbinom-and-rbinom-in-r/

https://rpubs.com/mpfoley73/458411

• Given a fixed number of trials - n

• For each trial there are two possible outcomes

• Probability of success remains same for all trials

• Trials are independent of each other

• We are interested in : Probability of x number of successes in n trials



**rbinom**

```r
#Probability of head in single coin flipping
rbinom(10,1,0.5)
```

```r
#Probability of head in 10 coin flipping
rbinom(10,10,0.5)
```

**pbinom**

```r
#Probability of 3 heads in 10 flips of coins
pbinom(3,10,0.5)
```

**qbinom**

```r
qbinom(0.17,10,0.5)
```

**dbinom**

```r
dbinom(3,10,0.5)
```

```r
#Draw the probability distribution of the case of flipping 10 coins and the looking at the probabilities of various number of heads.
n <- c(1:10)
(dbinom(n,10,0.5))


```

```r
barplot(dbinom(n,10,0.5))
```
